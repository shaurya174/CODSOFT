```python
#Get the data
train_df = pd.read_csv('fraudTrain.csv')
test_df = pd.read_csv('fraudTest.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
```

```python
encode.fit(train_df['merchant'])
```

```python
train_df['merchant']=encode.transform(train_df['merchant'])
```

```python
train_df
```

```python
# Separate fraud and non-fraud transactions
fraud_df = train_df[train_df['is_fraud'] == 1]
nonfraud_df = train_df[train_df['is_fraud'] == 0]

# Sample non-fraud transactions (3x number of fraud rows)
nonfraud_sample = nonfraud_df.sample(n=len(fraud_df) * 3, random_state=42)

# Combine fraud and sampled non-fraud data
balanced_df = pd.concat([fraud_df, nonfraud_sample])

# Shuffle the final dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```
